```python
# removing stopword and punctuation from tweets, and converting to lowercase
print('\033[92m')
print(all_tweets[:3])
print('\033[94m')
cleaned_tweets = []
for tweet in all_tweets:
    cleaned_tweet = []
    for word in tweet:
        if (word.lower() not in stop_words and word not in string.punctuation):
            cleaned_tweet.append(word.lower())
    cleaned_tweets.append(cleaned_tweet)
print('removed stop words and punctuation:')
print(cleaned_tweets[:3])
```

```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def predict_naive_bayes_proba(tweet_vectorized):
    tweet_vectorized = tweet_vectorized.ravel()  # ensure 1D

    # Calculate log-posteriors
    log_positive_posterior = np.log(positive_prior) + np.sum(np.log(positive_word_probs) * tweet_vectorized)
    log_negative_posterior = np.log(negative_prior) + np.sum(np.log(negative_word_probs) * tweet_vectorized)

    # Calculate log-odds and sigmoid
    z = log_positive_posterior - log_negative_posterior
    return sigmoid(z)  # returns probability of class 1

# Predict soft probabilities on test set
y_probs = np.array([predict_naive_bayes_proba(X_test_vectorized[i].toarray()[0]) for i in range(X_test_vectorized.shape[0])])

# Convert to binary class prediction (0 or 1) using a threshold, default 0.5
y_pred = (y_probs >= 0.5).astype(int)

# Evaluate
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Naive Bayes Accuracy (with sigmoid):", accuracy)

# Optional: Print first 5 probabilities
print("\nFirst 5 predicted probabilities:")
print(y_probs[:5])
```

```python
import numpy as np

def predict_prob(tweet_vectorized):
    tweet_vectorized = tweet_vectorized.ravel()
    log_pos = np.log(positive_prior) + np.sum(np.log(positive_word_probs) * tweet_vectorized)
    log_neg = np.log(negative_prior) + np.sum(np.log(negative_word_probs) * tweet_vectorized)

    # Convert from log-probabilities to probabilities using softmax trick
    max_log = max(log_pos, log_neg)
    log_pos -= max_log
    log_neg -= max_log
    prob_pos = np.exp(log_pos)
    prob_neg = np.exp(log_neg)
    total = prob_pos + prob_neg
    return prob_pos / total  # return P(y=1 | x)

# Calculate cost (log loss) over the test set
eps = 1e-9  # to avoid log(0)
log_losses = []
for i in range(X_test_vectorized.shape[0]):
    x = X_test_vectorized[i].toarray()[0]
    y_true = y_test[i]
    prob = predict_prob(x)
    prob = np.clip(prob, eps, 1 - eps)  # for numerical stability
    loss = - (y_true * np.log(prob) + (1 - y_true) * np.log(1 - prob))
    log_losses.append(loss)

avg_log_loss = np.mean(log_losses)
print("Naive Bayes Log Loss (Cost):", avg_log_loss)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score

# Step 1: Calculate Prior Probabilities P(y)
positive_prior = np.sum(y_train == 1) / len(y_train)  # Probability of positive class
negative_prior = np.sum(y_train == 0) / len(y_train)  # Probability of negative class

# Step 2: Calculate Likelihoods P(x|y) for each class
vocab_size = X_train_vectorized.shape[1]  # Number of unique words/features

# Apply Laplace smoothing (+1)
positive_word_counts = X_train_vectorized[y_train == 1].sum(axis=0) + 1
negative_word_counts = X_train_vectorized[y_train == 0].sum(axis=0) + 1

# Convert to 1D numpy arrays
positive_word_counts = np.asarray(positive_word_counts).flatten()
negative_word_counts = np.asarray(negative_word_counts).flatten()

# Total word count per class
positive_total_words = np.sum(positive_word_counts)
negative_total_words = np.sum(negative_word_counts)

# Probability of each word given the class
positive_word_probs = positive_word_counts / positive_total_words
negative_word_probs = negative_word_counts / negative_total_words

# Step 3: Define the prediction function using log probabilities
def predict_naive_bayes(tweet_vectorized):
    tweet_vectorized = tweet_vectorized.ravel()  # Ensure 1D shape

    log_positive_posterior = np.log(positive_prior) + np.sum(np.log(positive_word_probs) * tweet_vectorized)
    log_negative_posterior = np.log(negative_prior) + np.sum(np.log(negative_word_probs) * tweet_vectorized)

    return 1 if log_positive_posterior > log_negative_posterior else 0

# Step 4: Predict on test set
y_pred = np.array([
    predict_naive_bayes(X_test_vectorized[i].toarray()[0])
    for i in range(X_test_vectorized.shape[0])
])

# Step 5: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Naive Bayes Accuracy:", accuracy)

# Optional: Print first 5 predictions
print("\nFirst 5 predictions:", y_pred[:5])
print("First 5 true labels:", y_test[:5])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

# Simulate accuracy over 10 folds
kf = KFold(n_splits=10, shuffle=True, random_state=42)
accuracies = []

for train_index, val_index in kf.split(X_train_vectorized):
    X_train_k, X_val_k = X_train_vectorized[train_index], X_train_vectorized[val_index]
    y_train_k, y_val_k = y_train[train_index], y_train[val_index]

    # ---- Train Naive Bayes on fold ----
    # Prior probabilities
    pos_prior = np.sum(y_train_k == 1) / len(y_train_k)
    neg_prior = np.sum(y_train_k == 0) / len(y_train_k)

    # Word counts
    pos_wc = X_train_k[y_train_k == 1].sum(axis=0) + 1
    neg_wc = X_train_k[y_train_k == 0].sum(axis=0) + 1

    pos_wc = np.asarray(pos_wc).flatten()
    neg_wc = np.asarray(neg_wc).flatten()

    pos_total = pos_wc.sum()
    neg_total = neg_wc.sum()

    pos_probs = pos_wc / pos_total
    neg_probs = neg_wc / neg_total

    def predict(tweet_vec):
        tweet_vec = tweet_vec.ravel()
        log_pos = np.log(pos_prior) + np.sum(np.log(pos_probs) * tweet_vec)
        log_neg = np.log(neg_prior) + np.sum(np.log(neg_probs) * tweet_vec)
        return 1 if log_pos > log_neg else 0

    y_val_pred = np.array([predict(X_val_k[i].toarray()[0]) for i in range(X_val_k.shape[0])])
    acc = accuracy_score(y_val_k, y_val_pred)
    accuracies.append(acc)

# ---- Plot the "Accuracy vs Fold" ----
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), accuracies, marker='o', linestyle='-', color='green')
plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.title('Naive Bayes Accuracy over 10-Fold Cross Validation')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# STEP 1: Sigmoid helper
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# STEP 2: Probability prediction using sigmoid on log-odds
def predict_naive_bayes_proba(tweet_vectorized):
    tweet_vectorized = tweet_vectorized.ravel()

    log_positive_posterior = np.log(positive_prior) + np.sum(np.log(positive_word_probs) * tweet_vectorized)
    log_negative_posterior = np.log(negative_prior) + np.sum(np.log(negative_word_probs) * tweet_vectorized)

    z = log_positive_posterior - log_negative_posterior
    return sigmoid(z)

# STEP 3: Get probabilities and binary predictions
y_probs_test = np.array([predict_naive_bayes_proba(X_test_vectorized[i].toarray()[0]) for i in range(X_test_vectorized.shape[0])])
y_pred_test = (y_probs_test >= 0.5).astype(int)

y_probs_train = np.array([predict_naive_bayes_proba(X_train_vectorized[i].toarray()[0]) for i in range(X_train_vectorized.shape[0])])
y_pred_train = (y_probs_train >= 0.5).astype(int)

# STEP 4: Evaluation
train_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)

print(f"Naive Bayes Training Accuracy: {train_accuracy:.4f}")
print(f"Naive Bayes Test Accuracy: {test_accuracy:.4f}")

print("\nTraining Classification Report:\n", classification_report(y_train, y_pred_train))
print("Test Classification Report:\n", classification_report(y_test, y_pred_test))

# STEP 5: Confusion Matrices
train_conf_matrix = confusion_matrix(y_train, y_pred_train)
test_conf_matrix = confusion_matrix(y_test, y_pred_test)

print("\nTraining Confusion Matrix:\n", train_conf_matrix)
print("Test Confusion Matrix:\n", test_conf_matrix)

# STEP 6: Plotting Test Set Confusion Matrix
plt.figure(figsize=(7, 6))
sns.heatmap(test_conf_matrix, annot=True, fmt='d', cmap='Purples',
            xticklabels=["Pred 0", "Pred 1"], yticklabels=["Actual 0", "Actual 1"],
            cbar_kws={'label': 'Number of Predictions'})
plt.title("Naive Bayes Confusion Matrix - Test Set", fontsize=14)
plt.xlabel("Predicted", fontsize=12)
plt.ylabel("Actual", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```
